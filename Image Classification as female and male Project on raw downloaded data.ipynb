# Concatenate both male female datasets

```python
x=np.concatenate((x1,x2),axis=0)
x
```

# Generate labels for male female data

```python
y1=np.zeros(20)
y1
```

```python
y2=np.ones(20)
y2
```

```python
#concatenate y1 and y2
y=np.concatenate((y1,y2),axis=0)
y
```

# Splitting data into train test

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,shuffle=True,random_state=10)
```

```python
#standar scaller
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
sc.fit_transform(xtrain,ytrain)
```
